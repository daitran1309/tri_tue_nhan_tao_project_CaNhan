```python

# Trạng thái đích cần đạt được
GOAL = [1, 2, 3,
        8, 0, 4,
        7, 6, 5]    # 0 là ô trống

# ============================================================
# BƯỚC 1: Định nghĩa NODE (giống hệt BFS)
# ============================================================

def tao_node(state, duong_di, do_sau):
    node = {
        "state":    state,       # mảng 9 phần tử
        "duong_di": duong_di,    # đường đi từ đầu đến node này
        "do_sau":   do_sau       # độ sâu hiện tại (DFS cần biết)
    }
    return node


# ============================================================
# BƯỚC 2: Tìm các node con (giống hệt BFS)
# ============================================================

def tim_cac_node_con(node):
    state = node["state"]
    vi_tri_trong = state.index(0)

    hang = vi_tri_trong // 3
    cot  = vi_tri_trong % 3

    cac_node_con = []

    huong = [
        ("lên",    hang - 1, cot),
        ("xuống",  hang + 1, cot),
        ("trái",   hang,     cot - 1),
        ("phải",   hang,     cot + 1),
    ]

    for ten_huong, hang_moi, cot_moi in huong:
        if 0 <= hang_moi < 3 and 0 <= cot_moi < 3:
            vi_tri_moi = hang_moi * 3 + cot_moi

            state_moi = state.copy()
            state_moi[vi_tri_trong] = state_moi[vi_tri_moi]
            state_moi[vi_tri_moi]  = 0

            duong_di_moi = node["duong_di"] + [state_moi]
            node_con = tao_node(state_moi, duong_di_moi, node["do_sau"] + 1)
            cac_node_con.append(node_con)

    return cac_node_con


# ============================================================
# BƯỚC 3: In bảng 3x3
# ============================================================

def in_bang(state):
    for i in range(3):
        hang = state[i*3 : i*3+3]
        hang_str = []
        for so in hang:
            if so == 0:
                hang_str.append(" _")
            else:
                hang_str.append(f" {so}")
        print(" ".join(hang_str))
    print()


# ============================================================
# BƯỚC 4: Thuật toán DFS-Graph
#

#   Thêm giới hạn độ sâu (depth limit) để tránh lặp vô tận.
# ============================================================

def DFS_Graph(trang_thai_dau, trang_thai_dich, gioi_han_do_sau=30):

    print("=" * 45)
    print(f"BẮT ĐẦU DFS-Graph (giới hạn độ sâu = {gioi_han_do_sau})")
    print("=" * 45)

    # --- node ← NODE(problem.INITIAL) ---
    node = tao_node(trang_thai_dau, [trang_thai_dau], do_sau=0)

    # --- if GOAL-TEST(node.STATE) then return SOLUTION ---
    if node["state"] == trang_thai_dich:
        print("Trạng thái ban đầu đã là đích rồi!")
        return node["duong_di"]

    # --- frontier ← STACK ---
    # Dùng mảng (list) làm ngăn xếp LIFO
    # Thêm vào cuối (.append), lấy ra từ CUỐI (.pop())
    frontier = [node]

    # --- explored ← {} (tập rỗng) ---
    explored = []

    buoc = 0

    # --- while not EMPTY?(frontier) do ---
    while len(frontier) > 0:

        buoc += 1


        node = frontier[-1]
        frontier.pop()

        # --- explored ← explored ∪ {node.STATE} ---
        explored.append(node["state"])

        # In 10 bước đầu để minh họa
        if buoc <= 10:
            print(f"Bước {buoc}: độ sâu={node['do_sau']} | stack={len(frontier)} | explored={len(explored)}")

        # Nếu đạt giới hạn độ sâu thì không mở rộng tiếp
        if node["do_sau"] >= gioi_han_do_sau:
            continue

        # --- for each action in problem.ACTIONS(node.STATE) do ---
        cac_con = tim_cac_node_con(node)

        for child in cac_con:

            # --- if child.STATE ∉ explored AND child ∉ frontier then ---
            da_explored = child["state"] in explored
            da_frontier = any(f["state"] == child["state"] for f in frontier)

            if not da_explored and not da_frontier:

                # --- if GOAL-TEST(child.STATE) then return SOLUTION ---
                if child["state"] == trang_thai_dich:
                    print(f"...")
                    print(f"\nTÌM THẤY LỜI GIẢI sau {buoc} bước!")
                    print(f"Nodes đã explored: {len(explored)}")
                    print(f"Độ dài đường đi: {len(child['duong_di']) - 1} bước di chuyển\n")
                    return child["duong_di"]

                # --- frontier.INSERT(child) ---
                # DFS push vào CUỐI stack
                frontier.append(child)

    # --- return failure ---
    print(f"\nTHẤT BẠI: Không tìm thấy lời giải trong giới hạn độ sâu {gioi_han_do_sau}!")
    return None


# ============================================================
# CHẠY CHƯƠNG TRÌNH
# ============================================================

if __name__ == "__main__":

    trang_thai_dau = [2, 8, 3,
                      1, 6, 4,
                      7, 0, 5]

    print("Trạng thái ban đầu:")
    in_bang(trang_thai_dau)

    print("Trạng thái đích:")
    in_bang(GOAL)

    loi_giai = DFS_Graph(trang_thai_dau, GOAL, gioi_han_do_sau=30)

    if loi_giai:
        print("=" * 45)
        print("LỜI GIẢI TỪNG BƯỚC:")
        print("=" * 45)
        for i, state in enumerate(loi_giai):
            if i == 0:
                print(f"Bước {i} (ban đầu):")
            else:
                print(f"Bước {i}:")
            in_bang(state)
```
