```python
import random

# Cài đặt kích thước lưới
N = 4

def random_grid():
    # Tạo lưới N x N toàn số 0 và 1 bằng cách thủ công
    grid = []
    for i in range(N):
        row = []
        for j in range(N):
            row.append(random.randint(0, 1))
        grid.append(row)
    return grid

def print_grid(grid, r, c, step, action):
    # In thông tin bước hiện tại
    print(f"\nBước {step} | Hành động: {action} | Vị trí: ({r}, {c})")

    # In lưới ra màn hình
    for i in range(N):
        row_display = "| "
        for j in range(N):
            if i == r and j == c:
                row_display += "A | " # Agent (Người máy)
            else:
                row_display += str(grid[i][j]) + " | "
        print("-" * (N * 4 + 1))
        print(row_display)
    print("-" * (N * 4 + 1))

# --- KHỞI TẠO ---
grid = random_grid()
r = random.randint(0, N-1) # Dòng ngẫu nhiên
c = random.randint(0, N-1) # Cột ngẫu nhiên
step = 0

print_grid(grid, r, c, step, "BẮT ĐẦU")

# --- VÒNG LẶP CHÍNH ---
# Tiếp tục chạy khi vẫn còn ô có số 1 (rác)
while any(1 in row for row in grid):
    step += 1

    if grid[r][c] == 1:
        # Nếu có rác thì dọn (biến 1 thành 0)
        grid[r][c] = 0
        action = "DỌN DẸP"
    else:
        # Nếu sạch thì chọn hướng đi ngẫu nhiên
        possible_moves = []
        if r > 0:   possible_moves.append("LÊN")
        if r < N-1: possible_moves.append("XUỐNG")
        if c > 0:   possible_moves.append("TRÁI")
        if c < N-1: possible_moves.append("PHẢI")

        move = random.choice(possible_moves)

        if move == "LÊN":    r -= 1
        if move == "XUỐNG":  r += 1
        if move == "TRÁI":   c -= 1
        if move == "PHẢI":   c += 1
        action = f"DI CHUYỂN {move}"

    print_grid(grid, r, c, step, action)

print(f"\nHoàn thành! Tổng cộng mất {step} bước để dọn sạch.")
```
